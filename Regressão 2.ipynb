```python
#REGRESSÃO MULTIPLA
import numpy as np
import statsmodels.api as sm #estatiística
import sys
import matplotlib.pyplot as plt  
from matplotlib.animation import FuncAnimation
from mpl_toolkits.axes_grid.anchored_artists import AnchoredText
import csv
from openpyxl import load_workbook

def read_xlsx(name,L_n,inter):
    wb = load_workbook(name+'.xlsx')
    sheet = wb.active
    cut = sheet[inter][0][0].value ; string='p'
    for cell in sheet[inter]:
        if cell[0].value !=0 and type(cell[0].value)!=type(string):
            L_n.append(float(str(cell[0].value).replace(',', '.')))
    return cut

cond = 's'
while cond == 's':
    list1 =[] ; list2= [] ; list3= []
    tab = 'CompleteTTT3'
    cond1 = '2'
    if cond1 =='2':
        Label = []
        interval =  input('Intervalo das colunas. Ex:X1:X2;Z1:Z2;Y1:Y2 ')
        interval = interval.split(';') ; size = len(interval)
        lc=[list1,list2,list3]
        for i, L_n in enumerate(lc[0:size]):
            Label.append(read_xlsx(tab,L_n,interval[i]))
    else: break
            
    x = np.column_stack((list1,list2)) 

    x = sm.add_constant(x, prepend=True) #Adiciona a coluna das constantes

    res = sm.OLS(list3,x).fit() #Cria e ajusta o modelo

    print(res.params) #coef.

    print(res.bse) #erros

    print(res.summary())
    
    s=res.params
    eq=0
    eq1=0
    print('Resíduo')
    for k, Ln2 in enumerate(list3):
        eq=list3[k] - (s[0]+s[1]*list1[k]+s[2]*list2[k])
        eq1= eq1+(eq*eq)
        print(eq)
    RMS = (eq1/len(list3))**0.5
    print('\n'+'RMS='+str(RMS))

    cond = input('Nova regressão? [s/n] ')
    print('\n')
```
